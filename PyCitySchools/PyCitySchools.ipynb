```python
#Setup
import pandas as pd
from pathlib import Path
import numpy as np
```

```python
#Load Files
studt_data = Path("Resources/students_complete.csv")
schl_data = Path("Resources/schools_complete.csv")
```

```python
#Read Files
student_data = pd.read_csv(studt_data)
school_data = pd.read_csv(schl_data)
```

```python
school_final_data = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_final_data.head()
```

```python
## District Summary
```

```python
#Unique Schools
unique_schools = school_final_data['school_name'].unique()
print(unique_schools)
```

```python
#Unique School Total
school_count = len(unique_schools)
print(school_count)
```

```python
#Student Total
student_count = school_final_data['student_name'].count()
print(student_count)
```

```python
#School Budgets
unique_schools_budget = school_final_data['budget'].unique()
print(unique_schools_budget)
```

```python
#Total School Budgets
total_budget = unique_schools_budget.sum()
print(total_budget)
```

```python
#Math Scores
average_math = school_final_data['math_score'].mean()
print(average_math)
```

```python
#Reading Scores
average_read = school_final_data['reading_score'].mean()
print(average_read)
```

```python
#Passed Math (>= %70)
pass_math = school_final_data[(school_final_data["math_score"]>=70)].count()["student_name"]
pass_math_percent = pass_math / float(student_count)*100
print(pass_math_percent)
```

```python
#Passed Reading (>= %70)
pass_read = school_final_data[(school_final_data["reading_score"]>=70)].count()["student_name"]
pass_read_percent = pass_read / float(student_count)*100
print(pass_read_percent)
```

```python
#Passes Reading and Math
pass_read_math = school_final_data[(school_final_data["math_score"]>=70) & (school_final_data["reading_score"]>=70)].count()["student_name"]
pass_read_math_percent = pass_read_math / float(student_count) *100
print(pass_read_math_percent)
```

```python
#Summary
district_summary = {'Total Schools': [school_count],'Total Students': [student_count], 'Total Budget': [total_budget], 'Average Math Score': [average_math],'Average Reading Score': [average_read],'% Passing Math': [pass_math_percent],'% Passing Reading': [pass_read_percent],'% Overall Passing': [pass_read_math_percent]}

district_summary_df = pd.DataFrame(district_summary)

district_summary_df["Total Students"] = district_summary_df["Total Students"].map("{:,}".format)
district_summary_df["Total Budget"] = district_summary_df["Total Budget"].map("${:,.2f}".format)

district_summary_df
```

```python
## School Summary
```

```python
#Set index to type per school from school_data
school_type = school_data.set_index(["school_name"])["type"]
school_type
```

```python
#Student Counts per School
student_countsper = student_data.groupby('school_name')['student_name'].count()
print(student_countsper)
```

```python
#Calculate school budget 
unique_schools_budgetper = school_data.groupby('school_name')['budget'].sum()
print(unique_schools_budgetper)
```

```python
#Calculate per capita/student spending
unique_schools_percapita = unique_schools_budgetper / student_countsper
print(unique_schools_percapita)
```

```python
#Average Math Test Scores Per School
average_math = student_data.groupby('school_name')['math_score'].mean()
print(average_math)
```

```python
#Average Reading Test Scores Per School
average_read = student_data.groupby('school_name')['reading_score'].mean()
print(average_read)
```

```python
#Students Passed Math (>= %70) per School
pass_mathper = student_data[(student_data["math_score"]>=70)].groupby('school_name')["student_name"].count()
pass_mathper_percent = pass_mathper / student_countsper*100
print(pass_mathper_percent)
```

```python
#Students Passed Reading (>= %70) per School
pass_readper = student_data[(student_data["reading_score"]>=70)].groupby('school_name')["student_name"].count()
pass_readper_percent = pass_readper / student_countsper*100
print(pass_readper_percent)
```

```python
#Students Passed Reading and Math (>= %70) per School
pass_read_mathper = student_data[(student_data["math_score"]>=70) & (student_data["reading_score"]>=70)].groupby('school_name')["student_name"].count()
pass_read_mathper_percent = pass_read_mathper / student_countsper *100
print(pass_read_mathper_percent)
```

```python
#Data Frame Summary

school_summary_df = pd.DataFrame({'School Type' : school_type, 'Total Students' : student_countsper, 'Total School Budget' : unique_schools_budgetper, 'Per Student Budget' : unique_schools_percapita,
                                  'Average Math Score' : average_math, 'Average Reading Score' : average_read, '% Passing Math' : pass_mathper_percent, '% Passing Reading' : pass_readper_percent,
                                  '% Overall Passing' : pass_read_mathper_percent
    
})

school_summary_df["Per Student Budget"] = school_summary_df["Per Student Budget"].map("${:,.2f}".format)
school_summary_df["Total School Budget"] = school_summary_df["Total School Budget"].map("${:,.2f}".format)

school_summary_df
```

```python
#Highest - Performing Schools (by% Overall Passing)
```

```python
top_schools = school_summary_df.sort_values(by='% Overall Passing', ascending=False)
top_schools.head()
```

```python
#Bottom - Performing Schools (by% Overall Passing)
```

```python
bottom_schools = school_summary_df.sort_values(by='% Overall Passing', ascending=True)
bottom_schools.head()
```

```python
#Math Scores By Grade
```

```python
#Different Grades
ninth_gr = student_data[(student_data["grade"]=="9th")]
tenth_gr = student_data[(student_data["grade"]=="10th")]
eleventh_gr = student_data[(student_data["grade"]=="11th")]
twelfth_gr = student_data[(student_data["grade"]=="12th")]

ninth_math_score = ninth_gr.groupby('school_name')['math_score'].mean()
tenth_math_score = tenth_gr.groupby('school_name')['math_score'].mean() 
eleventh_math_score = eleventh_gr.groupby('school_name')['math_score'].mean()
twelfth_math_score = twelfth_gr.groupby('school_name')['math_score'].mean()

math_scores_grade_df = pd.DataFrame({'9th':ninth_math_score, '10th':tenth_math_score, '11th':eleventh_math_score, '12th': twelfth_math_score})
math_scores_grade_df.index.name = None
math_scores_grade_df
```

```python
#Reading Score By Grade
```

```python
#Different Grades
ninth_gr = student_data[(student_data["grade"]=="9th")]
tenth_gr = student_data[(student_data["grade"]=="10th")]
eleventh_gr = student_data[(student_data["grade"]=="11th")]
twelfth_gr = student_data[(student_data["grade"]=="12th")]

ninth_read_score = ninth_gr.groupby('school_name')['reading_score'].mean()
tenth_read_score = tenth_gr.groupby('school_name')['reading_score'].mean() 
eleventh_read_score = eleventh_gr.groupby('school_name')['reading_score'].mean()
twelfth_read_score = twelfth_gr.groupby('school_name')['reading_score'].mean()

reading_scores_grade_df = pd.DataFrame({'9th':ninth_read_score, '10th':tenth_read_score, '11th':eleventh_read_score, '12th': twelfth_read_score})
reading_scores_grade_df.index.name = None
reading_scores_grade_df
```

```python
#Scores By School Spending
```

```python
#Bins
spending_bins = [0,585,640,645,680]
labels = ["<$585", "$585-630", "$630-645", "$645-680"]
```

```python
#New School Summary
school_spending_df = school_summary_df.copy()
school_spending_df["Per Student Budget"] = school_spending_df["Per Student Budget"].replace('[\$,]', '', regex=True).astype(float)
school_spending_df["Per Student Budget"] = school_spending_df["Per Student Budget"].astype(int)
print(school_spending_df["Per Student Budget"].dtype)
```

```python
#Categorize Spending
school_spending_df["Spending Ranges (Per Student)"] = pd.cut(school_spending_df["Per Student Budget"], bins=spending_bins, labels=labels)
school_spending_df["Per Student Budget"] = school_spending_df["Per Student Budget"].map("${:,.2f}".format)
school_spending_df
```

```python
#Averages
spending_math_scores = school_spending_df.groupby(["Spending Ranges (Per Student)"])["Average Math Score"].mean()
spending_reading_scores = school_spending_df.groupby(["Spending Ranges (Per Student)"])["Average Reading Score"].mean()
spending_passing_math = school_spending_df.groupby(["Spending Ranges (Per Student)"])["% Passing Math"].mean()
spending_passing_reading = school_spending_df.groupby(["Spending Ranges (Per Student)"])["% Passing Reading"].mean()
overall_passing_spending = school_spending_df.groupby(["Spending Ranges (Per Student)"])["% Overall Passing"].mean()

spending_reading_scores
```

```python
#Create into One Data Frame
spending_summary_df = pd.DataFrame({'Average Math Score': spending_math_scores, 'Average Reading Score': spending_reading_scores, '% Passing Math': spending_passing_math,
                                   '% Passing Reading': spending_passing_reading, '% Overall Passing': overall_passing_spending}
                                   )
spending_summary_df
```

```python
#Scores by School Size
```

```python
#Make Bins
size_bins = [0,1000,2000,5000]
labels_2 = ["Small (<1000)", "Medium (1000-2000)", "Large (2000-5000)"]
```

```python
#Categorize the spending
school_summary_df["School Size"] = pd.cut(school_summary_df["Total Students"], bins=size_bins, labels=labels_2)
school_summary_df
```

```python
#Averages
size_math_scores = school_summary_df.groupby(["School Size"])["Average Math Score"].mean()
size_reading_scores = school_summary_df.groupby(["School Size"])["Average Reading Score"].mean()
size_passing_math = school_summary_df.groupby(["School Size"])["% Passing Math"].mean()
size_passing_reading = school_summary_df.groupby(["School Size"])["% Passing Reading"].mean()
size_overall_passing = school_summary_df.groupby(["School Size"])["% Overall Passing"].mean()
```

```python
#Create into One Data Frame
size_summary_df = pd.DataFrame({'Average Math Score': size_math_scores, 'Average Reading Score': size_reading_scores, '% Passing Math': size_passing_math,
                                   '% Passing Reading': size_passing_reading, '% Overall Passing': size_overall_passing}
                                   )
size_summary_df
```

```python
#Scores by School Type
```

```python
#Average by School Type
average_math_score_by_type = school_summary_df.groupby(["School Type"])["Average Math Score"].mean()
average_reading_score_by_type = school_summary_df.groupby(["School Type"])["Average Reading Score"].mean()
average_percent_passing_math_by_type = school_summary_df.groupby(["School Type"])["% Passing Math"].mean()
average_percent_passing_reading_by_type = school_summary_df.groupby(["School Type"])["% Passing Reading"].mean()
average_percent_overall_passing_by_type = school_summary_df.groupby(["School Type"])["% Overall Passing"].mean()
```

```python
#New Data Frame
type_summary_df = pd.DataFrame({'Average Math Score': average_math_score_by_type, 'Average Reading Score': average_reading_score_by_type, '% Passing Math': average_percent_passing_math_by_type,
                                   '% Passing Reading': average_percent_passing_reading_by_type, '% Overall Passing': average_percent_overall_passing_by_type}
                                   )
type_summary_df
```
